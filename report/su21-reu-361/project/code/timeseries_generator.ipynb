```python
from pandas import read_csv
from matplotlib import pyplot
series = read_csv("C:/Users/Sledgehammer/Desktop/eos.csv", header=0, index_col=0, parse_dates=True, squeeze=True)
print(series.head())
```

```python
series.drop("Open", axis=1, inplace=True)
series.drop("High", axis=1, inplace=True)
series.drop("Low", axis=1, inplace=True)
series.drop("Volume", axis=1, inplace=True)
series.drop("Market Cap", axis=1, inplace=True)
```

```python
series.plot()
pyplot.title("EOS Price Over the Years")
pyplot.xlabel("Date")
pyplot.ylabel("EOS Price")
pyplot.tight_layout()
pyplot.savefig("C://Users//Sledgehammer//Desktop//eos_price.png", dpi=100, facecolor="#FFFFFF")
```
